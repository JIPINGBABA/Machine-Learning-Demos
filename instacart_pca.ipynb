```python
import pandas as pd
```

```python
#1、get dataset
order_products=pd.read_csv("./instacart/order_products__prior.csv")
products=pd.read_csv("./instacart/products.csv")
orders=pd.read_csv("./instacart/orders.csv")
aisles=pd.read_csv("./instacart/aisles.csv")
```

```python
products
```

2.Merge table
order_products__prior.csv：Order and product information

fields：order_id, product_id, add_to_cart_order, reordered
products.csv：product information

fields：product_id, product_name, aisle_id, department_id
orders.csv：The user's order information
fields：order_id,user_id,eval_set,order_number,….
aisles.csv：The specific item category to which the product belongs
fields： aisle_id, aisle

```python
#merge aisles and products aisle and product_id
tab1=pd.merge(aisles,products,on=["aisle_id","aisle_id"])
```

```python
tab1
```

```python
tab2=pd.merge(tab1,order_products,on=["product_id","product_id"])
```

```python
tab2
```

```python
tab3=pd.merge(tab2,orders,on=["order_id","order_id"])
```

```python
tab3
```

```python
#3、Find the relationship between user_ID and aisle
table=pd.crosstab(tab3["user_id"],tab3["aisle"])
```

```python
table
```

```python
data=table[:10000]
```

```python
#4、 PCA dimensionality reduction
from sklearn.decomposition import PCA
```

```python
#1、Instantiate a transformer class
transfer=PCA(n_components=0.95)
#2、call fit_transform
data_new=transfer.fit_transform(data)
```
